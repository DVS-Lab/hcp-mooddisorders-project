```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
explore = pd.read_csv('exploratory_roi.csv')
```

```python
explore.head()
```

```python
explore["DEPMENT"].mean()
```

```python
%config InlineBackend.figure_format = 'svg'

# data to plot
n_groups = 3 #depresion, family history, then HCs
mental_means = (explore["DEPMENT"].mean(), explore["FAMMENT"].mean(), explore["HCMENT"].mean())
random_means = (explore["DEPRAND"].mean(), explore["DEPRAND"].mean(), explore["DEPRAND"].mean())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = .55

mentalbars = plt.bar(index, mental_means, bar_width,
alpha=opacity,
edgecolor="#000000",
linewidth=1,
color='b',
yerr=(explore["DEPMENT"].std()/math.sqrt(71), explore["FAMMENT"].std()/math.sqrt(103), explore["HCMENT"].std()/math.sqrt(105)),
capsize=4,
label='Ment')

randombars = plt.bar(index + bar_width, random_means, bar_width,
alpha=opacity,
edgecolor="#000000",
linewidth=1,
color='purple',
yerr=(explore["DEPRAND"].std()/math.sqrt(71), explore["FAMRAND"].std()/math.sqrt(103), explore["HCRAND"].std()/math.sqrt(105)),
capsize=4,
label='Rand')

plt.xlabel('Group')
plt.ylabel('Timecourse')
plt.title('ROI Data')
plt.xticks(index + bar_width-0.2, ('Dep', 'Fam', 'HC'))
plt.legend()
plt.axhline(linewidth=1, color='black')

plt.tight_layout()
plt.show()
```

```python
%config InlineBackend.figure_format = 'svg'

# data to plot
n_groups = 3 #depresion, family history, then HCs
mental_means = (explore["DEPMENT"].mean(), explore["FAMMENT"].mean(), explore["HCMENT"].mean())
random_means = (explore["DEPRAND"].mean(), explore["DEPRAND"].mean(), explore["DEPRAND"].mean())

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 1

mentalbars = plt.bar(index, mental_means, bar_width,
alpha=opacity,
edgecolor="#000000",
linewidth=1,
color='b',
yerr=(explore["DEPMENT"].std()/math.sqrt(71), explore["FAMMENT"].std()/math.sqrt(103), explore["HCMENT"].std()/math.sqrt(105)),
capsize=4,
label='Ment')

randombars = plt.bar(index + bar_width, random_means, bar_width,
alpha=opacity,
edgecolor="#000000",
linewidth=1,
color='purple',
yerr=(explore["DEPRAND"].std()/math.sqrt(71), explore["FAMRAND"].std()/math.sqrt(103), explore["HCRAND"].std()/math.sqrt(105)),
capsize=4,
label='Rand')

plt.tight_layout()

plt.axis('On')
plt.axhline(linewidth=1, color='black')
ax.set_xticklabels([])
ax.set_yticklabels([])
plt.tight_layout()

plt.savefig("exploratory_roi.svg",format='svg', transparent=True)
plt.show()
```
