```python
import pandas as pd
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
roi = pd.read_csv('Fam-HC_MentPPI_roiresults.csv')
```

```python
roi.head()
```

```python
fam_ment_avg = roi["FAMMENT"].mean()
fam_ment_std = roi["FAMMENT"].std()
fam_ment_se = (fam_ment_std)/(math.sqrt(130))

dep_ment_avg = roi["DEPMENT"].mean()
dep_ment_std = roi["DEPMENT"].std()
dep_ment_se = (dep_ment_std)/(math.sqrt(71))

hc_ment_avg = roi["HCMENT"].mean()
hc_ment_std = roi["HCMENT"].std()
hc_ment_se = (hc_ment_std)/(math.sqrt(105))

fam_rand_avg = roi["FAMRAND"].mean()
fam_rand_std = roi["FAMRAND"].std()
fam_rand_se = (fam_rand_std)/(math.sqrt(103))

dep_rand_avg = roi["DEPRAND"].mean()
dep_rand_std = roi["DEPRAND"].std()
dep_rand_se = dep_rand_std/(math.sqrt(71))

hc_rand_avg = roi["HCRAND"].mean()
hc_rand_std = roi["HCRAND"].std()
hc_rand_se = hc_rand_std/math.sqrt(105)
```

```python
print('Mental_PPI Data:')
print('fam_ment_avg=',fam_ment_avg,'dep_ment_avg=',dep_ment_avg,'hc_meant_avg=',hc_ment_avg)
print('fam_ment_se=',fam_ment_se,'dep_ment_se=',dep_ment_se,'hc_meant_se=',hc_ment_se)
print('')
print('Random_PPI Data:')
print('fam_rand_avg=',fam_rand_avg,'dep_rand_avg=',dep_rand_avg,'hc_rand_avg=',hc_rand_avg)
print('fam_rand_se=',fam_rand_se,'dep_rand_se=',dep_rand_se,'hc_rand_se=',hc_rand_se)
```

```python
# data to plot
n_groups = 3 #depresion, family history, then HCs
#need to change to HC, Depression, FAM
mental_means = (hc_ment_avg,dep_ment_avg,fam_ment_avg)
random_means = (hc_rand_avg,dep_rand_avg,fam_rand_avg)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.55

mentalbars = plt.bar(index, mental_means, bar_width,
alpha=opacity,
color='b',
yerr=(hc_ment_se,dep_ment_se,fam_ment_se),
capsize=4,
label='Ment')

randombars = plt.bar(index + bar_width, random_means, bar_width,
alpha=opacity,
color='purple',
yerr=(hc_rand_se,dep_rand_se,fam_rand_se),
capsize=4,
label='Rand')

plt.xlabel('Group')
plt.ylabel('Timecourse')
plt.title('ROI Data')
plt.xticks(index + bar_width-0.2, ('HC', 'Dep', 'Fam'))
plt.legend()
plt.axhline(linewidth=1, color='black')

plt.tight_layout()
plt.show()
```
