```python
import pandas as pd
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
roi = pd.read_csv('/Users/lindseytepfer/Desktop/fslmeants_roi.csv')
```

```python
roi.head()
```

```python
fam_ment_avg = roi["fam_ment"].mean()
fam_ment_std = roi["fam_ment"].std()
fam_ment_se = (fam_ment_std)/(math.sqrt(130))

dep_ment_avg = roi["dep_ment"].mean()
dep_ment_std = roi["dep_ment"].std()
dep_ment_se = (dep_ment_std)/(math.sqrt(71))

hc_ment_avg = roi["hc_ment"].mean()
hc_ment_std = roi["hc_ment"].std()
hc_ment_se = (hc_ment_std)/(math.sqrt(105))

fam_rand_avg = roi["fam_rand"].mean()
fam_rand_std = roi["fam_rand"].std()
fam_rand_se = (fam_rand_std)/(math.sqrt(103))

dep_rand_avg = roi["dep_rand"].mean()
dep_rand_std = roi["dep_rand"].std()
dep_rand_se = dep_rand_std/(math.sqrt(71))

hc_rand_avg = roi["hc_rand"].mean()
hc_rand_std = roi["hc_rand"].std()
hc_rand_se = hc_rand_std/math.sqrt(105)
```

```python
print('Mental_PPI Data:')
print('fam_ment_avg=',fam_ment_avg,'dep_ment_avg=',dep_ment_avg,'hc_meant_avg=',hc_ment_avg)
print('fam_ment_se=',fam_ment_se,'dep_ment_se=',dep_ment_se,'hc_meant_se=',hc_ment_se)
print('')
print('Random_PPI Data:')
print('fam_rand_avg=',fam_rand_avg,'dep_rand_avg=',dep_rand_avg,'hc_rand_avg=',hc_rand_avg)
print('fam_rand_se=',fam_rand_se,'dep_rand_se=',dep_rand_se,'hc_rand_se=',hc_rand_se)
```

```python
# data to plot
n_groups = 3 #depresion, family history, then HCs
mental_means = (-0.013625291366197185, 0.010943515087378642, -0.03866278411428572)
random_means = (0.0014339421971830985, 0.010485282912621357, -0.0307549891142857)

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.55

mentalbars = plt.bar(index, mental_means, bar_width,
alpha=opacity,
color='b',
yerr=(0.00964670823499975,0.006576638821755958,0.008603011702734208),
capsize=4,
label='Ment')

randombars = plt.bar(index + bar_width, random_means, bar_width,
alpha=opacity,
color='purple',
yerr=(0.006601535084550757,0.0048329580745265915,0.005572499224234995),
capsize=4,
label='Rand')

plt.xlabel('Group')
plt.ylabel('Timecourse')
plt.title('ROI Data')
plt.xticks(index + bar_width-0.2, ('Dep', 'Fam', 'HC'))
plt.legend()
plt.axhline(linewidth=1, color='black')

plt.tight_layout()
plt.show()
```
